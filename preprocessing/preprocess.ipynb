```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../csi_dataset/csi_labeled_20260209_140928.csv")

print(df.shape)
print(df)
```

```python
amp_cols = [c for c in df.columns if c.startswith("amp_")]
X = df[amp_cols].astype(float)
t = df["timestamp"].astype(float) 
```

```python
plt.figure()
plt.plot(t, X)  # matplotlib will plot each column as a separate line
plt.xlabel("Time (packet index)")
plt.ylabel("Amplitude")
plt.title("All subcarriers as lines over time (raw)")
plt.show()
```
